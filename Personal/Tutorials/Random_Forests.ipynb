```python
import pandas as pd
import numpy as np
from IPython.display import display
```

<h4>Initialize dataset</h4>

```python
dataset = pd.read_csv('airfoil_self_noise.dat', sep='\s+', header=None)
dataset.rename(columns={
    0: 'Frequency (Hz)', 
    1: 'Angle (deg)', 
    2: 'Chord length (m)', 
    3: 'Free-stream velocity (m/s)', 
    4: 'Suction side displacement thickness (m)',
    5: 'Scaled sound pressure level (dB)'
}, inplace=True)
display(dataset)
```

<h4>Split dataset for targeting</h4>

```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .25, random_state = 2022)

randomForestModel = RandomForestRegressor(max_depth=3)
randomForestModel.fit(X_train, Y_train)

print(randomForestModel.score(X_test, Y_test))
print(randomForestModel.score(X_train, Y_train))

decisionTreeModel = DecisionTreeRegressor(max_depth=3)
decisionTreeModel.fit(X_train, Y_train)

print(decisionTreeModel.score(X_test, Y_test))
print(decisionTreeModel.score(X_train, Y_train))
```
